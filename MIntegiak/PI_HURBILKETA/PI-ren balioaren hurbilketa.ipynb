### Azalpena: 

Mintegi honetan zorizko zenbakiak sortuz, Pi zenbakia hurbildu daitekeela ikusiko dugu.

Beheko irudian dagoen zirkunferentziaren erradioa bat bada, gune berdearen azalera pi/4=0.78539816339 da

Mintegi honetan N zorizko zenbaki sortuko dugu $(0,1)\times(0,1)$ karratuan eta, gune berdearen barruan daudenen proportzioa hartuko dugu pi/4-ren estimatzailetzat.

#### Prozedura


Talde bakoitzak algoritmo bat egingo dugu, ondokoa egingo duelarik. Hasieran, kontadore=0 balioa ezarri eta
N aldiz, ondokoa egingo luke algoritmoak:
* U(0,1) banaketa jarraitzen duen zenbaki bat sortu eta x deitu
* U(0,1) banaketa jarraitzen duen zenbaki bat sortu eta y deitu
* (x,y) bektorea gune berdean badago, kontadore=kontadore+1. 

pi/4-ren estimatzailetzat kontadore/N hartuko dugu.

#### Egin beharrekoa

Jarduera honetan pi/4 balioaren estimazio azterketa egin behar da. Lehendabizi, goiko algoritmoa programatu behar da (Graduan ikasitako edozein programa erabiliz). Ondoren, N balioaren arabera lortutako estimazioaren kalitatea (edo errorea) aztertu behar da. Horrela:

* N txikia den kasuan, zenbait irudi aurkeztu behar dira. Irudi hauek erakutsi behar dute estimatzaileak erraz kalkula daitezkela grafikoa errepatuz eta lortutako balioak ezberdinak izan daitezkela (eta, gainera, guztiak pi/4-ren baliotik urrun daudela)

* N balioa handiagoak hartuta, ondoko galderak erantzun behar dira: zer gertatzen da lortutako estimatzaileen aldakortasunarekin? Zein da N-ren balioa estimatzailearen errorea gehienez 0.1 izan dadin? Eta errorea 0.01 izateko? 

Jarduera honetan bikoteka lan egingo duzue eta bi mintegi saioetan burutuko da. Garrantzitsua da jarduera honetako lana klasean egitea, hau da, ez da onartuko klasetik kanpo egindako lana. Horrela, bigarren saioaren bukaeran lana bukatu ez dute taldeek, luzapen txiki bat eskatu ahal dute. 

Jarduera honetan egindako lana txostena batean idatzi behar da. Txostena eta egindako kodea eGela bidez bidaliko da irakasleari bigarren saioaren bukaeran (edo luzapenaren azken momentuan).
